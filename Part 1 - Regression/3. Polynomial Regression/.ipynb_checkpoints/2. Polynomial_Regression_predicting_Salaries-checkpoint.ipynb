```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
# importing dataset
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
dataset.head()
```

```python
x = dataset.iloc[:, 1:-1].values
x
```

```python
y = dataset.iloc[:, -1].values
y
```

```python
# Visualising the x and y
plt.figure(figsize=(10,6))
plt.scatter(x,y, s=15)
plt.xlabel('Predictor - Years', fontsize=16)
plt.ylabel('Target - Salaries', fontsize=16)
plt.show()
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
```

```python
# Training the Model
lm = LinearRegression()
lm.fit(x.reshape(-1,1), y.reshape(-1,1))
```

```python
# Let see how the linear regression performs on the non-linear data
y_pred = lm.predict(x.reshape(-1,1))
```

```python
# predicting the new value using Linear Regression
y_pred = lm.predict([[6.5]])
```

```python
print("Predicted Salary for 6.5 years using Linear Model is => ", y_pred)
```

```python
# Let see how the linear regression performs on the non-linear data
y_hat = lm.predict(x.reshape(-1,1))
```

```python
# Plotting the predicted value with the actual value
plt.figure(figsize=(10,6))
plt.scatter(x, y, s=15)
plt.plot(x, y_hat, color='r')
plt.xlabel('Predictor - Years', fontsize=16)
plt.ylabel('Target - Salaries', fontsize=16)
plt.show()
```

```python
# importing library for calculating mean_squared error
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error_mse = mean_squared_error(y, y_hat)
root_mean_square_error_rmse = np.sqrt(mean_squared_error_mse)
```

```python
print('MSE for Linear Regression => ', mean_squared_error_mse)
print('RMSE for Linear Regression => ', root_mean_square_error_rmse)
```

   <b> The RMSE and MSE values are too high. </b>

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
# for degree 4
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(x)
lm_2 = LinearRegression()
lm_2.fit(x_poly, y)
```

<b>Visualising the Polynomial Regression results</b>

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, lm_2.predict(poly_reg.fit_transform(x)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
y_ht = lm_2.predict(poly_reg.fit_transform([[6.5]]))
print("Predicted Salary for 6.5 years using Poly Linear Model is => ",y_ht)
```

```python
# for degree 3
poly_reg = PolynomialFeatures(degree = 3)
x_poly = poly_reg.fit_transform(x)
lm_2 = LinearRegression()
lm_2.fit(x_poly, y)

plt.scatter(x, y, color = 'red')
plt.plot(x, lm_2.predict(poly_reg.fit_transform(x)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
y_ht = lm_2.predict(poly_reg.fit_transform([[6.5]]))
print("Predicted Salary for 6.5 years using Poly Linear Model is => ",y_ht)
```

```python
# for degree 2
poly_reg = PolynomialFeatures(degree = 2)
x_poly = poly_reg.fit_transform(x)
lm_2 = LinearRegression()
lm_2.fit(x_poly, y)

plt.scatter(x, y, color = 'red')
plt.plot(x, lm_2.predict(poly_reg.fit_transform(x)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
y_ht = lm_2.predict(poly_reg.fit_transform([[6.5]]))
print("Predicted Salary for 6.5 years using Poly Linear Model is => ",y_ht)
```

<b> predicted value for the 6.5 years for degree 4 is much closer to the future salary <b/>
